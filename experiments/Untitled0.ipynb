```python
from sklearn.neighbors import KernelDensity
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10,20))

for i, n_samples in enumerate(np.linspace(100, 20000, 20)):
    kde = KernelDensity(
        bandwidth=.2,
        kernel='gaussian', algorithm='ball_tree')
    kde.fit(np.random.normal(size=(n_samples,1)))

    y = np.linspace(-10,10,200)[:, np.newaxis]
    x = np.exp(kde.score_samples(y))

    plt.subplot(20, 1, i+1)
    ax = plt.plot(y, x)
    plt.gca().set_xlim((-10, 10))
```
